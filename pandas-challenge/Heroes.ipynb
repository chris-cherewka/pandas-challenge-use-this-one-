```python
import pandas as pd
```

```python
file = "Resources/purchase_data.csv"
data_df = pd.read_csv(file)
data_df.head()
```

```python
#Player Count

total_players = len(data_df.SN.value_counts())
#total_players

#create new data frame
Players_Count = pd.DataFrame({"Total Players":[total_players]
})
Players_Count
```

```python
unique_items = len(data_df["Item Name"].unique())
#unique_items

#find total number of items and sum of prices
total_items = data_df["Item Name"].count()
total_items = total_items.astype(float)
sum_price = data_df.Price.sum()
sum_price = sum_price.astype(float)

average_price = sum_price / total_items
#print(type(average_price))

average_price = "${:.2f}".format(average_price)
#print(average_price)

total_purch = len(data_df["Purchase ID"])
#total_purch

revenue = data_df.Price.sum()
revenue = "${:.2f}".format(revenue)
#revenue
#print(revenue)

#create new data frame
Purchasing_Analysis = pd.DataFrame([{
    "Number of Unique Items": unique_items,
    "Average Price": average_price,
    "Number of Purchases": total_purch,
    "Total Revenue": revenue   
}])

Purchasing_Analysis
```

```python
#Gender Demographics

#group data by Gender
gender = data_df.groupby(["Gender"])

total_gender = gender.nunique()["SN"]

percent = total_gender / total_players

percent = percent.astype(float).map("{:.2%}".format)
percent

gender_demographics = pd.DataFrame({
    "Total Count": total_gender,
    "Percentage of Players": percent     
})

gender_demographics
```

```python
#Gender Purchasing Analysis

#use grouped data from previous analysis
purchase_count = gender["Purchase ID"].count()
#purchase_count

avg_purchase_price = gender.Price.mean()
avg_purchase_price = avg_purchase_price.astype(float).map("${:.2f}".format)
#avg_purchase_price

purchase_value = gender.Price.sum()
#purchase_value = purchase_value.astype(float).map("${:.2f}".format)
#purchase_value

avg_purchase_person = purchase_value/total_gender
avg_purchase_person =avg_purchase_person.astype(float).map("${:.2f}".format)
#avg_purchase_person
purchase_value = purchase_value.astype(float).map("${:.2f}".format)

Purchasing_Analysis_Gender = pd.DataFrame({
    "Purchase Count": purchase_count,
    "Average Purchase Price": avg_purchase_price,
    "Total Purchase Value": purchase_value,
    "Avg Total Purchase per Person": avg_purchase_person
})

Purchasing_Analysis_Gender
```

```python
#Age Demographics

bins = [0, 9, 14, 19, 24, 29, 34, 39, 100]
names = ['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+']

data_df["Age Group"] = pd.cut(data_df["Age"], bins, labels=names)
#data_df

age_data = data_df.groupby("Age Group")

#total number of unique players in age group
total_age = age_data.SN.nunique()
#total_age

#percentage of each age group in data
perc_age = round(total_age / total_players * 100, 2)
#perc_age

#format percentages
perc_age = perc_age.astype(float).map("{:.2f}%".format)
#perc_age

#create data frame
Age_Demographics = pd.DataFrame({
    "Total Count": total_age,
    "Percentage of Players": perc_age,
})

#Age_Demographics
#drop index
Age_Demographics.index.name = None
Age_Demographics

```

```python
# need to create Purchase Count, Average Purchase Price, Total Purchase Value, and Avg Total Purchase per Person

#finding average purchase coutn
age_purchase_count = age_data["Purchase ID"].count()
#age_purchase_count

#find average purchase price
avg_purchase_price_age = age_data.Price.mean()
#avg_purchase_price_age

#Total Purchase Value
total_purchase_value_age = age_data.Price.sum()
#total_purchase_value_age

#Average Total purchase per person
avg_total_purchase_person = total_purchase_value_age / total_age
#avg_total_purchase_person




#create data frame and format data
Age_Purchasing_Analysis =  pd.DataFrame({
    "Purchase Count": age_purchase_count,
    "Average Purchase Price": avg_purchase_price_age,
    "Total Purchase Value": total_purchase_value_age,
    "Avg Total Purchase per Person": avg_total_purchase_person
})

Age_Purchasing_Analysis
```

```python
#Top Spenders

#calculating total purchase count per each person
spenders_df = data_df.groupby("SN")
spenders_purchase_count = spenders_df["Purchase ID"].count()
#spenders_purchase_count


#calculating average purchase price per each person
spenders_avg_purchase_price = spenders_df.Price.mean()
#spenders_avg_purchase_price

#calculating total purchase value for each person
spenders_total_purchase_value = spenders_df.Price.sum()
#spenders_total_purchase_value


#create new data frame
Top_Spenders_Analysis =  pd.DataFrame({
    "Purchase Count": spenders_purchase_count,
    "Average Purchase Price": spenders_avg_purchase_price,
    "Total Purchase Value": spenders_total_purchase_value,
})

#Sort data
Top_Spenders_Analysis = Top_Spenders_Analysis.sort_values(["Total Purchase Value"], ascending=False)


Top_Spenders_Analysis.head()
```

```python
#Most Popular Items

items = data_df[["Item ID", "Item Name", "Price"]]
#items

#group data
items_df = items.groupby(["Item ID", "Item Name"])

items_purchase_count = items_df["Item Name"].count()
#items_purchase_count

items_total_purchase_value = items_df.Price.sum()
#items_total_purchase_value

item_price = items_total_purchase_value / items_purchase_count
#item_price



Most_Popular_Items = pd.DataFrame({"Purchase Count": items_purchase_count, 
                                   "Item Price": item_price,
                                   "Total Purchase Value": items_total_purchase_value
})

Most_Popular_Items = Most_Popular_Items.sort_values(["Purchase Count"], ascending=False)

Most_Popular_Items.style.format({"Item Price": "${:,.2f}",
                                "Total Purchase Value": "${:,.2f}"})

Most_Popular_Items.head()
```

```python
#Most Profitable

Most_Popular_Items = Most_Popular_Items.sort_values(["Total Purchase Value"], ascending=False)

Most_Popular_Items.head()
```
